```python
#Write a Python program to add 'ing' at the end of a given string (length should be at least 3).
#If the given string already ends with 'ing' then add 'ly' instead If the string length of the given
#string is less than 3, leave it unchanged 


def A1(string):
    Length = len(string)

    if Length > 2:
        if string[-3] == "ing":
            string += "ly"
        else:
            string += "ing"

    return string
print(A1("ab"))
print(A1("abc"))
print(A1("string"))
```

```python
#Write a Python program to find the first appearance of the substring 'not' and 'poor' from a given string, if
#'not' follows the 'poor', replace the whole 'not'...'poor' substring with 'good'.
#Return the resulting string 

def B1(string2):
    snot = string2.find("not")
    spoor = string2.find("poor")



    if spoor > snot and snot>0 and spoor>0:
        string2 = string2.replace(string2[snot:(spoor+4)], "good")
        return string2
    else:
        return string2
print(B1("The lyrics is not that poor !"))
print(B1("The lyrics are poor !"))
```

```python
#Program to find Greatest Common Divisor of two numbers.
#For example, the GCD of 20 and 28 is 4 and GCD of 98 and 56 is 14.
def gcd(a,b):
    result = min(a, b)
   
    while result:
      if a % result == 0 and b % result == 0:
        break
      result -= 1
   
  
    return result
 

a = 98
b = 56
 
print(f"GCD of {a} and {b} is {gcd(a, b)}")
```

```python
#Write a Python program to check whether a list contains a sublist.
test_list = [5,6,3,8,2,1,7,1]

print("The original list : " + str(test_list))

sub_list = [8,2,1]

res = False
for idx in range(len(test_list) - len(sub_list) + 1):
        if test_list[idx : idx + len(sub_list)] == sub_list:
            res = True
            break

print("Is sub_list present in list ? : " + str(res))
```

```python
#Write a Python program to sort a dictionary (ascending /descending) by value.
#Ascending
d={
    'c':3,
    'd':4,
    'a':1,
    'e':5,
    'b':2
}
print(d)
for i in sorted(d,key=d.get):
    print(i,d[i])

#descending
d={
    'c':2,
    'd':5,
    'a':3,
    'e':4,
    'b':1
}
print(d)
for i in sorted(d,key=d.get,reverse=True):
    print(i,d[i])
```

```python
#Write a Python program to find the highest 3 values in a dictionary
from collections import Counter


dictionary = {
              'A' : 67, 'B' : 23, 'C' : 45,
              'D' : 56, 'E' : 12, 'F' : 69}
D = Counter(dictionary)

highest = D.most_common(3)

print("Initial Dictionary : ")
print(dictionary, "\n")

print("Dictionary with 3 highest values : ")
print("Keys : values")

for i in highest:
    print(i[0], " :",i[1]," ")
    
```

```python
#Given a number n, write a python program to make and print the list of Fibonacci series up to n.
#Input : n=7
#Hint : first 7 numbers in the series
#Expected output :
#First few Fibonacci numbers are 0, 1, 1, 2, 3, 5, 8, 13

class fibonacci:
    def series(self,S):
        num1,num2 = 0,1
        print("Fibonacci series : ",num1,num2, end = " ")

        for i in range(2,S):
            num3 = num1 + num2
            num1 = num2
            num2 = num3
            print(num3, end=" ")

obj = fibonacci()
obj.series(5)
```

```python
#Python Program to Find Factorial of Number Using Recursion
def factorial(n):
    if (n==1 or n==0):

        return 1
    else:
        return (n * factorial(n - 1))

num = 5;
print("number : ",num)
print("Factorial : ",factorial(num))        
```
